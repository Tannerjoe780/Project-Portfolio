## 8.  Define
Create two columns that use the best prediction and dog breed

## Code

```python
# start by making our two new columns
breed = []
best_conf_lvl=[]

def optimize_pred(p_cleanv1):
    if p_cleanv1.p1_dog == True:
        breed.append(p_cleanv1.p1)
        best_conf_lvl.append(p_cleanv1.p1_conf)
    elif p_cleanv1.p2_dog == True:
        breed.append(p_cleanv1.p2)
        best_conf_lvl.append(p_cleanv1.p2_conf)
    elif p_cleanv1.p3_dog == True:
        breed.append(p_cleanv1.p3)
        best_conf_lvl.append(p_cleanv1.p3_conf)
    else:
        breed.append('Unknown')
        best_conf_lvl.append(0)
        
#create our new column based on the conditions above
        
p_cleanv1.apply(optimize_pred, axis = 1)
p_cleanv1['breed'] = breed
p_cleanv1['best_conf_lvl']= best_conf_lvl

#save this new column
```

### Issue #1: The twitter archive data has four columns doggo, floofer, pupper, and puppo are all stages of dog and should be in one column.
